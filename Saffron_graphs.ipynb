### Conversi√≥n de datos de columna a lista e implementaci√≥n de running average

```python
saffron_data = pd.read_csv("desktop/Saffron_36300_785.txt", delimiter="\t", header=None,usecols=[0,1], names=['x', 'y'])

x = saffron_data['x'].to_list() #to convert data in a col to list
y = saffron_data['y'].to_list()

def movingaverage (values, window):
    weights = np.repeat(1.0, window)/window
    sma = np.convolve(values, weights, 'valid')
    return sma

yMA = movingaverage(y,3)
#print yMA

plt.figure(figsize=(5, 3), dpi=300)

plt.title("Saffron")
plt.xlabel(r'raman shift ($cm^{-1}$)')
plt.ylabel("intensity")

plt.plot(x[len(x)-len(yMA):],yMA)
plt.show()
    

#

#
#
#plt.plot(x, y)
#plt.show()  #
```
